Handling dan Remove Outlier

```python
# Memilih kolom-kolom yang ingin dimasukkan ke dalam DataFrame baru 'data'
data = df[['date','open_price', 'first_trade', 'high', 'low', 'close', 'change', 'volume']]
data.head(10)
```

```python


import pandas as pd
import datetime
# Convert the 'date' column to datetime format
data['date'] = pd.to_datetime(data['date'])
# Filter the DataFrame to only include rows after October 12, 2021
data = data[data['date'] > datetime.datetime(2021, 10, 12)]
```

```python
data.head(10)
```

```python
# Memilih kolom-kolom (objek)
cat_cols = data.select_dtypes(include=['object']).columns
# Memilih kolom-kolom numerik
num_cols = data.select_dtypes(include=np.number).columns.tolist()

print(cat_cols)
print(num_cols)
```

```python
# Membuat boxplot untuk setiap kolom numerik
for col in num_cols:
  plt.title(col)
  sns.boxplot(x=data[col])  # Menggambar boxplot dengan seaborn
  plt.show()
```

```python
  # Drop data outliers
data = data.drop(data[data['change'] < -400].index)
data = data.drop(data[data['volume'] > 300000000].index)
```

```python
# Melihat kolom change setelah menghapus outlier
plt.title('change')
sns.boxplot(x=data['change'])
plt.show()
```

```python
# melihat jumlah barus data outliers pada kolom volume
data.loc[data['volume'] > 240000000]
```

```python
# Drop data outliers
data = data.drop(data[data['volume'] > 240000000].index)
```

```python
# Melihat kolom volume setelah menghapus outlier
plt.title('volume')
sns.boxplot(x=data['volume'])
plt.show()
```

```python
X = data.loc[:, ['open_price', 'first_trade', 'high', 'low', 'volume']]
Y = data['close']

print(X)
print(Y)
```

# Percobaan 1 20% 80%

```python
# Split the data into training and testing sets
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.20, random_state=10)
```

```python
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression


model = LinearRegression().fit(X_train, Y_train)
intercept_model = model.intercept_
coefficient_model = model.coef_

print(intercept_model)
print(coefficient_model)
```

y prediksi = konstanta + koefisien 1 + koefisien 2 + dst

```python
#Training Set
Y_pred = model.predict(X_train)

data_actual = Y_train[0:15].values
hasil_prediksi = Y_pred[0:15]

df_trainset = pd.DataFrame({
    'Data Actual': data_actual,
    'Hasil Prediksi': hasil_prediksi
})
df_trainset
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(Y_train, Y_pred, label='Train Data')
plt.plot([Y.min(), Y.max()], [Y.min(), Y.max()], color='red', lw=2)
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title('Actual vs Predicted (Train Data)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
index = np.arange(len(Y_train[0:10]))

plt.figure(figsize=(10, 6))

# Plot actual values
plt.plot(index, Y_train[0:10], label='Actual', marker='o')

# Plot predicted values
plt.plot(index, Y_pred[0:10], label='Predicted', marker='x')

plt.xlabel('Index')
plt.ylabel('Close Price')
plt.title('Actual vs Predicted Close Prices (Train Data)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Testing Set
Y_pred_test = model.predict(X_test)

data_actual = Y_test[0:15].values
hasil_prediksi = Y_pred_test[0:15]

df_testingset = pd.DataFrame({
    'Data Actual': data_actual,
    'Hasil Prediksi': hasil_prediksi
})
df_testingset
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(Y_test, Y_pred_test, label='Test Data')
plt.plot([Y.min(), Y.max()], [Y.min(), Y.max()], color='red', lw=2)
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title('Actual vs Predicted (Test Data)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
index = np.arange(len(Y_train[0:10]))

plt.figure(figsize=(10, 6))

# Plot actual values
plt.plot(index, Y_test[0:10], label='Actual', marker='o')

# Plot predicted values
plt.plot(index, Y_pred_test[0:10], label='Predicted', marker='x')

plt.xlabel('Index')
plt.ylabel('Close Price')
plt.title('Actual vs Predicted Close Prices')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Train Set
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error

rmse = mean_squared_error(Y_train, Y_pred, squared=False)
mape = mean_absolute_percentage_error(Y_train, Y_pred)
mse = mean_squared_error(Y_train, Y_pred)
r2 = r2_score(Y_train, Y_pred)

print("Root Mean Squared Error:", rmse)
print("Mean Absolute Percentage Error:", mape)
print("Mean Squared Error:", mse)
print("R-squared:", r2)
```

```python
#Test Set
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error

rmse = mean_squared_error(Y_test, Y_pred_test, squared=False)
mape = mean_absolute_percentage_error(Y_test, Y_pred_test)
mse = mean_squared_error(Y_test, Y_pred_test)
r2 = r2_score(Y_test, Y_pred_test)

print("Root Mean Squared Error:", rmse)
print("Mean Absolute Percentage Error:", mape)
print("Mean Squared Error:", mse)
print("R-squared:", r2)
```
